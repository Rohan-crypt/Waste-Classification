```python
import tensorflow as tf

# If they downloaded the file manually:
model = tf.keras.models.load_model("waste_classifier_mobilenetv2.h5")
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing import image
from google.colab import files

uploaded = files.upload()  # upload any test image

for fn in uploaded.keys():
    img_path = fn
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    # Prediction
    pred = model.predict(img_array)[0][0]
    label = "E-waste" if pred > 0.5 else "Non-E-waste"

    # Show result
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Prediction: {label} ({pred:.2f})")
    plt.show()
```
